**Dataset Overview**

1. Price: The selling price of the house.
2. Area: The total area (in square feet) of the house.
3. Bedrooms: The number of bedrooms in the house.
4. Bathrooms: The number of bathrooms in the house.
5. Stories: The number of floors/stories in the house.
6. Mainroad: Whether the house is located on a main road (yes or no).
7. Guestroom: Whether the house has a guest room (yes or no).
8. Basement: Whether the house has a basement (yes or no).
9. Hotwaterheating: Whether the house has hot water heating (yes or no).
10. Airconditioning: Whether the house has air conditioning (yes or no).
11. Parking: The number of parking spaces available with the house.
12. Prefarea: Whether the house is in a preferred area (yes or noo).
13. Furnishingstatus: The furnishing status of the house (Furnished, Semi-furnished, Unfurnished).

**Import Neccesary Libraries**

**Question 1: What is the relationship between the price of a house and its area?**

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Area', y='Price', data=df)
plt.title('House Price vs Area')
plt.xlabel('Area (square feet)')
plt.ylabel('Price')
plt.show()
```

###### The plot illustrates that there is a positive correlation between the area of a house and its price. As the area increases, the price of the house tends to increase as well. This suggests that larger houses generally command higher prices.

**Question 2: How does the number of bedrooms and bathrooms influence the price of a house?**

```python
# Plotting Price vs Bedrooms
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
sns.boxplot(x='Bedrooms', y='Price', data=df, linewidth=2)
plt.title('House Price vs Number of Bedrooms')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Price')

# Plotting Price vs Bathrooms
plt.subplot(1, 2, 2)
sns.boxplot(x='Bathrooms', y='Price', data=df,linewidth=2)
plt.title('House Price vs Number of Bathrooms')
plt.xlabel('Number of Bathrooms')
plt.ylabel('Price')

plt.show()
```

**Bedrooms**: There's a general trend that houses with more bedrooms tend to have higher prices. However, the relationship is not strictly linear, and there's considerable variation in price within each bedroom category.

**Bathrooms**: Similar to bedrooms, houses with more bathrooms tend to be priced higher. The price increase appears more pronounced with an increase in the number of bathrooms compared to bedrooms, indicating that bathrooms might have a stronger influence on house price.

**Question 3: Is there a correlation between the number of stories in a house and its price?**

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Stories', y='Price', data=df)
plt.title('House Price vs Number of Stories')
plt.xlabel('Number of Stories')
plt.ylabel('Price')
plt.show()
```

Analysis: The boxplot illustrates that there is a trend where houses with more stories tend to have higher prices.

**Question 4: Does the presence of amenities such as a mainroad, guestroom, basement, hot water heating, and air conditioning affect the house price?**

```python
amenities = ['Mainroad', 'Guestroom', 'Basement', 'Hotwaterheating', 'Airconditioning'] # Careating Array for all the Amenities

# Creating subplots for each amenity
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(14, 18))
axes = axes.flatten()

for i, amenity in enumerate(amenities):
    sns.boxplot(x=amenity, y='Price', data=df, ax=axes[i])
    axes[i].set_title(f'House Price vs {amenity}')
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Price')

plt.show()
```

Analysis: These plots indicate that the presence of certain amenities, such as a main road, guestroom, basement, hot water heating, and air conditioning, can have a positive impact on the price of a house.

**Question 5: Does the furnishing status of a house (furnished, semi-furnished, unfurnished) correlate with its price?**

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Furnishingstatus', y='Price', data=df)
plt.title('House Price vs Furnishing Status')
plt.xlabel('Furnishing Status')
plt.ylabel('Price (INR)')
plt.show()
```

The boxplot indicates that the furnishing status of a house does correlate with its price.

**Conclusion:**
The analysis highlights the importance of house size, number of facilities, amenities, and furnishing status in determining the price of a house. These factors are crucial for sellers to consider when pricing their homes and for buyers to consider when assessing the value of a potential home purchase.
